# ðŸ’¬ Alcon Compliance Chat Interface

This cell provides an interactive chatbot for exploring and improving marketing material.

### **Available Commands**

- **`/grade <text>`**  
  Runs a full compliance evaluation on the provided marketing text.  
  Loads policies for the selected product and updates the chatbot's long-term memory.

- **Normal Messages**  
  Type any question or message to continue the conversation with the chatbot.  
  The assistant will:
  - Explain compliance issues  
  - Suggest fixes  
  - Compare the last two compliance checks  
  - Use short-term and long-term memory to stay consistent  

- **Exit**  
  Type **`exit`**, **`quit`**, or **`bye`** to end the chat session.

### **Session Notes**

- The chatbot remembers your last **10 messages**.
- The compliance engine remembers the **current** and **previous** compliance results.
- Fix suggestions update automatically after each `/grade` command.

---

When you're ready, scroll down and start chatting.


```python
print("âœ¨ Chatbot ready! Type **exit** to stop.")
print("Use `/grade <text>` to run a new compliance check.")

while True:
    message = input("You: ")

    # Exit command
    if message.lower() in ["exit", "quit", "bye"]:
        display(Markdown("> **Chat ended.**"))
        break

    # Run new compliance check
    if message.startswith("/grade "):
        new_text = message[len("/grade "):].strip()

        if not new_text:
            display(Markdown("ðŸŸ¦ **Bot:** Please provide text after `/grade`."))
            continue
        
        compliance_grade = ctrl.run_compliance_check(new_text)

        display(Markdown(
            f"""
ðŸŸ¦ **You:** {new_text}

ðŸŸ© **Compliance Updated!**  
{json.dumps(compliance_grade, indent=2)}
            """
        ))
        continue


    # ---------------------------
    #  Normal chat
    # ---------------------------
    response = ctrl.chat_bot_message(message)

    display(Markdown(
        f"""
ðŸŸ¦ **You:** {message}

â¬œ **Bot:**  
{response}
        """
    ))

```
